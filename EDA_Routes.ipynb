# Data Analysis on City of Orillia Roads Dataset

## 1. Import Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2. Load the dataset

```python
file_path = 'City of Orllia_Roads by Class 1.csv'
df = pd.read_csv(file_path)
```

# 3. Display the First Few Rows of the Dataset

```python
print("First few rows of the dataset:")
print(df.head())
```

# 4. Display Basic Information About the Dataset

```python
print("\nBasic Information:")
print(df.info())
```

# 5. Display Summary Statistics for Numerical Columns

print("\nSummary Statistics (Numerical Features):")
print(df.describe())

# 6. Check for Missing Values

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

# 7. Check for Duplicate Rows

```python
print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

# 8. Visualize the Distribution of Numerical Features

```python
numerical_features = df.select_dtypes(include=[np.number]).columns

for feature in numerical_features:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

# 9. Visualize the Correlation Matrix for Numerical Features

```python
plt.figure(figsize=(12, 8))
corr_matrix = df[numerical_features].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# 10. Generate a Word Cloud of Street Names and Intersections

```python
from wordcloud import WordCloud

# Combine the text columns to create a single text input for the word cloud
text = ' '.join(df['ST_NAME'].tolist() + df['ST_INT'].tolist())

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
